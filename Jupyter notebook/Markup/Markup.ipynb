# Подготовка данных

1. Необходимо указать соотвутствующий тип данных для каждой колонки, согласно ее содержанию.
2. Необнодимо переименовать колонки на английский текст, для дальнейшей удобной обработки.

```python
# Чтение датайфрейма
df = pd.read_csv(
    filepath_or_buffer=path_file,                                        # путь к файлу
    skiprows=1,                                                          # игнорирование строки с русскими названиями колонок
    encoding='utf-8',                                                    # кодировка 
    sep=',',                                                             # разделитель столбцов
    decimal=',',                                                         # Указываем, что десятичный разделитель - запятая
    names=[
        'date_sale',                                                     # новые имена колонок
        'cheque', 
        'nomenclature', 
        'darcode', 
        'product_group',
        'characteristic',
        'waight',
        'total_amount',
        'cost_price'
          ],
    parse_dates=['date_sale'],                                           # указываем колонки с типами данныд "дата"
    date_format='%m/%d/%y %H:%M',                                        # указываем формат чтения даты.
    dtype={                                                              # явное указание типов данных
            'cheque': 'int',                                             
            'nomenclature': 'string', 
            'darcode': 'int',
            'product_group': 'string',
            'characteristic': 'string',
            'waight': 'float',
            'total_amount': 'float',
            'cost_price': 'float'
          }
)
```

# Фильтрация данных в соответствии с условиями.

1. Период анализа - с 1.04.23 по 10.04.23
2. В анализе должны участвовать только продажи изделий с общей суммой и себестоимостью больше 0 рублей

```python
df.tail(3)
```

```python
filtered_df = df.query(
    "(date_sale >= @pd.to_datetime('2023-04-01')) & "  # Дата >= 1 апреля 2023
    "(date_sale <= @pd.to_datetime('2023-04-10')) & "  # Дата <= 10 апреля 2023
    "(total_amount > 0) & "                            # Общая сумма > 0
    "(cost_price > 0)"                                 # Себестоимость > 0
)
```

```python
filtered_df
```

# Нахождение средней наценки для двух групп металлов.

Для группировки данных по металлу создам колонку, в которой будет указан метал проданного изделия. Исходные данные возьму из колонки "characteristic". Использовать буду регулярные выражения.

```python
filtered_df.loc[:, 'metal'] = filtered_df['characteristic'].str.extract(r'([A-Za-z]+)')[0]
```

```python
filtered_df
```

Для нахождения средней наценки, определю ее для каждого проданного товара, потом сгруппирую данные по металлую.

```python
filtered_df.loc[:, 'markup'] = filtered_df['total_amount'] - filtered_df['cost_price']
```

```python
filtered_df.head()
```

```python
result = (
    filtered_df
    .groupby(['metal'])  # Группировка по колонке 'metal'
    .agg(
        average_markup=('markup', 'mean')  # Вычисление среднего значения наценки
    )
    .round(1)  # Округление до 2 знаков после запятой
    .reset_index()  # Преобразование индекса в колонку
)
```

```python
result
```

```python
print(f'''Cредняя наценка на ювелирные изделия из золота составляет {result.average_markup[1]} рубля, из серебра - {result.average_markup[0]} рубля.''')
```
